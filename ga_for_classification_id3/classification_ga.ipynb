```python
def readFileCSV(filename):
    datas = []
    with open(filename) as csvfile:
        read_file = csv.reader(csvfile, delimiter=',')
        for row in read_file:
            datas.append(row)
    return datas

def encodeDataTrainToBinary(datas):
    data = []
    for row in datas:
        g = [0 for i in range (15)]
        if(row[0] == 'rendah'):
            g[0] = 1
            g[1] = 0
            g[2] = 0
        elif(row[0] == 'normal'):
            g[0] = 0
            g[1] = 1
            g[2] = 0
        elif(row[0] == 'tinggi'):
            g[0] = 0
            g[1] = 0
            g[2] = 1

        if(row[1] == 'pagi'):
            g[3] = 1
            g[4] = 0
            g[5] = 0
            g[6] = 0
        elif(row[1] == 'siang'):
            g[3] = 0
            g[4] = 1
            g[5] = 0
            g[6] = 0
        elif(row[1] == 'sore'):
            g[3] = 0
            g[4] = 0
            g[5] = 1
            g[6] = 0
        elif(row[1] == 'malam'):
            g[3] = 0
            g[4] = 0
            g[5] = 0
            g[6] = 1

        if(row[2] == 'berawan'):
            g[7] = 1
            g[8] = 0
            g[9] = 0
            g[10] = 0
        elif(row[2] == 'cerah'):
            g[7] = 0
            g[8] = 1
            g[9] = 0
            g[10] = 0
        elif(row[2] == 'hujan'):
            g[7] = 0
            g[8] = 0
            g[9] = 1
            g[10] = 0
        elif(row[2] == 'rintik'):
            g[7] = 0
            g[8] = 0
            g[9] = 0
            g[10] = 1

        if(row[3] == 'rendah'):
            g[11] = 1
            g[12] = 0
            g[13] = 0
        elif(row[3] == 'normal'):
            g[11] = 0
            g[12] = 1
            g[13] = 0
        elif(row[3] == 'tinggi'):
            g[11] = 0
            g[12] = 0
            g[13] = 1

        if(row[4] == 'ya'):
            g[14] = 1
        elif(row[4] == 'tidak'):
            g[14] = 0  
            
        data.append(g)
    return data
```

```python
def generateChromosome(szChrom):
    chrom = []  
    for i in range(szChrom):
        chrom.append(random.randint(0,1))      
    
    return chrom


def generatePopulation(szPop):
    pop = []
    szChrom = 30
    for i in range(szPop):
        pop.append(generateChromosome(szChrom))
        
    return pop
```

```python
def checkValid(ch, data, j, k):
    if((ch[j][0] == 1 and ch[j][0] == data[k][0]) or (ch[j][1] == 1 and ch[j][1] == data[k][1]) or (ch[j][2] == 1 and ch[j][2] == data[k][2])):
        temp = 1
    else :
        temp = 0
    if((ch[j][3] == 1 and ch[j][3] == data[k][3]) or (ch[j][4] == 1 and ch[j][4] == data[k][4]) or (ch[j][5] == 1 and ch[j][5] == data[k][5]) or (ch[j][6] == 1 and ch[j][6] == data[k][6])):
        time = 1
    else :
        time = 0
    if((ch[j][7] == 1 and ch[j][7] == data[k][7]) or (ch[j][8] == 1 and ch[j][8] == data[k][8]) or (ch[j][9] == 1 and ch[j][9] == data[k][9]) or (ch[j][10] == 1 and ch[j][10] == data[k][10])):
        cond = 1
    else :
        cond = 0
    if((ch[j][11] == 1 and ch[j][11] == data[k][11]) or (ch[j][12] == 1 and ch[j][12] == data[k][12]) or (ch[j][13] == 1 and ch[j][13] == data[k][13])):
        humid = 1
    else :
        humid = 0
    return (temp,time,cond,humid)
```

```python
 def calculateFitnessValue(chromo, data):
    valList = []
    ch = [chromo[i:i + 15] for i in range(0, len(chromo), 15)]
    
    for k in range(len(data)):
        nTrue = 0
        nFalse = 0
        
        for j in range(len(ch)):                
            valid = checkValid(ch, data, j, k)    

            if (valid[0] == 1 and valid[1] == 1 and valid[2] == 1 and valid[3] == 1):
                if (ch[j][14] == data[k][14]):
                    fVal = 1
                else:
                    fVal = 0
            else:
                if (ch[j][14] != data[k][14]):
                    fVal = 1
                else:
                    fVal = 0
            
            if (fVal == 1):
                nTrue += 1
                nFalse += 0
            else:
                nTrue += 0
                nFalse += 1
                
        if (nTrue > nFalse):
            valList.append(1)
        else:
            valList.append(0)
                    
    fitVal = 0        
    for b in range(len(valList)):
        if (valList[b] == 1):
            fitVal += 1
                    
    return fitVal/80
```

```python
def calculateAllFitnessValue(pop, data):
    fitnessValues = []
    for i in range(len(pop)):
        fitnessValues.append(calculateFitnessValue(pop[i], data))
    return fitnessValues
```

```python
def rouletteWheelSelection(pop, fit, data):
    total = 0
    for i in range(len(fit)):
        total += fit[i]
    
    r = random.random()
    krom = 0
    while (r > 0):
        r -= calculateFitnessValue(pop[krom], data)/total
        krom += 1
        
    return krom-1
```

```python
def crossover(par1, par2, pop, pc):
    #SEARCHING POINT
    rand = random.random()
    if (rand < pc):
        if (len(par2) < len(par1)):
            temp = copy.deepcopy(par1)
            par1 = copy.deepcopy(par2)
            par2 = copy.deepcopy(temp)
        p1 = random.randint(0,len(par1)-1)
        p2 = random.randint(0,len(par1)-1)
        while (p1 == p2):
            p2 = random.randint(0,len(par1)-1)
        pA = min(p1, p2)
        pB = max(p1, p2)

        lenPoint = abs(pA - pB)
        gap = lenPoint % 15

        pCa = pA + lenPoint
        pCb = pA + gap
        pDa = pB - lenPoint
        pDb = pB - gap

        pointTwo = []
        if (pA != pCa):
            pointTwo.append([pA, pCa])
        if (pA != pCb):
            pointTwo.append([pA, pCb])
        if (pDa != pB):
            pointTwo.append([pDa, pB])
        if (pDb != pB):
            pointTwo.append([pDb, pB])

        choosenPoint = random.randint(0,len(pointTwo)-1)
        
        #CROSSOVER PROCESS
        listA = par1[pA:pB]
        listB = par2[pointTwo[choosenPoint][0]:pointTwo[choosenPoint][1]]
        par1[pA:pB] = listB
        par2[pointTwo[choosenPoint][0]:pointTwo[choosenPoint][1]] = listA
    
    return (par1, par2)
```

```python
def mutation(chrom, pm):
    for i in range(len(chrom)):
        rand = random.random()
        if (rand < pm):
            if (chrom[i] == 0):
                chrom[i] = 1
            else:
                chrom[i] = 0
    return chrom
```

```python
def getElitism(fit_all):
    return fit_all.index(max(fit_all))
```

```python
def predictResult(data, chBest):
    ch = [chBest[i:i + 15] for i in range(0, len(chBest), 15)]

    for k in range(len(data)):
        nTrue = 0
        nFalse = 0
        for j in range(len(ch)):
            valid = checkValid(ch, data, j, k)    
            if (valid[0] == 1 and valid[1] == 1 and valid[2] == 1 and valid[3] == 1):
                nTrue += 1
                t = j
                nFalse += 0
            else:
                nTrue += 0
                f = j
                nFalse += 1 

        if (nTrue >= nFalse):
            data[k][14] = ch[t][14]
        else:
            if (ch[f][14] == 1):
                data[k][14] = 0
            else :
                data[k][14] = 1
                
    return data
    
```

```python
#HYPERPARAMETER

szPop = 30
generation = 500
pc = 0.76
pm = 0.17
```

```python
#GENERAL REPLACEMENT

file = 'data_train.csv'
dataTrain = readFileCSV(file)
data = encodeDataTrainToBinary(dataTrain)
population = generatePopulation(szPop)
for i in range(generation):
    fitnessAll = calculateAllFitnessValue(population, data)
    new_population = []

    best = getElitism(fitnessAll)
    new_population.append(population[best])
    new_population.append(population[best])
    i = 0
    while (i < szPop-2):
        par1 = rouletteWheelSelection(population, fitnessAll, data)
        par2 = rouletteWheelSelection(population, fitnessAll, data)
        while (par1 == par2):
            par2 = rouletteWheelSelection(population, fitnessAll, data)
        parent1 = copy.deepcopy(population[par1])
        parent2 = copy.deepcopy(population[par2])
        child = crossover(parent1, parent2, population, pc)
        mut1 = mutation(child[0],pm)
        mut2 = mutation(child[1],pm)
        new_population.append(mut1)
        new_population.append(mut2)
        i += 2
    population = new_population

result = getElitism(fitnessAll)
chromBest = population[result]

print('Kromosom terbaik:', chromBest) 
```

```python
fileTest = 'data_validation.csv'
dataUji = readFileCSV(fileTest)
dataU = encodeDataTestToBinary(dataUji)
predict = predictResult(dataU, chromBest )
```

```python
#SAVE INTO CSV FILE
pr = [[] for j in range(len(predict))]
for i in range(len(predict)):
    if (predict[i][14] == 1):
        pr[i] = ('ya')
    else:
        pr[i]= ('tidak')
with open('target_latih.csv', mode='w') as employee_file:
    employee_writer = csv.writer(employee_file, delimiter='\n', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    employee_writer.writerow(pr)
```
